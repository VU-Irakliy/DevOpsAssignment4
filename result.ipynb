```python
import os
os.system('pip install matplotlib')
import matplotlib.pyplot as plt

# Data extracted from the message logs
data = {
    "1 replica": {"Req/Sec": "0.89", "Min": "40.79", "Max": "0.98k", "Stdev": "76.17%"},
    "2 replicas": {"Req/Sec": "1.14", "Min": "227.26", "Max": "1.78k", "Stdev": "77.67%"},
    "4 replicas": {"Req/Sec": "1.06", "Min": "197.88", "Max": "1.34k", "Stdev": "57.42%"},
    "8 replicas": {"Req/Sec": "1.06", "Min": "180.04", "Max": "1.33k", "Stdev": "61.67%"},
}

# Prepare data for plotting
replicas = list(data.keys())
req_sec = [float(value["Req/Sec"]) * 1000 for key, value in data.items()]  # Convert k to actual number

# Create histogram
plt.figure(figsize=(10, 6))
plt.bar(replicas, req_sec, color='skyblue')
plt.xlabel('Number of Instances')
plt.ylabel('Requests per Second')
plt.title('Requests per Second by Number of Instances')
plt.show()
```
